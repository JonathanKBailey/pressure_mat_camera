```python
def camera_date_reader(d):
    try:
        return datetime.datetime.strptime(d, '%Y-%m-%dT%H_%M_%S.%f%z')
    except ValueError:
        return datetime.datetime.strptime(d, '%Y-%m-%dT%H_%M_%S%z')
```
